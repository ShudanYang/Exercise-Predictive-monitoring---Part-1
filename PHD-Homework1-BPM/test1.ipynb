```python
# 导入必要的库
import pandas as pd

# 1. 导入事件日志作为Pandas DataFrame
event_log_df = pd.read_csv('hospital_billing.csv')

# 2. 重命名属性为"caseid, activity, ts"（如果名称不同）
event_log_df = event_log_df.rename(columns={'case_id': 'caseid', 'activity': 'activity', 'timestamp': 'ts'})
#print(event_log_df)

# 3. 创建一个事件日志，仅保留caseid、activity和ts属性
event_log_new_df = event_log_df[['caseid', 'activity', 'ts']]

# 4. 创建一个新的DataFrame，包含caseid和按时间戳排序的事件列表
new_df = pd.DataFrame(columns=['caseid', 'list_of_events_ordered_by_timestamp'])

current_case = None#跟踪当前处理的案例
events = []#事件列表

# 按caseid和ts排序事件日志
event_log_df = event_log_df.sort_values(by=['caseid', 'ts'])

# 遍历事件日志的每一行
for index, row in event_log_df.iterrows():
    case_id = row['caseid']
    activity = row['activity']
    timestamp = row['ts']

    # 如果是新案例或当前案例已结束，创建新案例
    if case_id != current_case:
        if current_case is not None:
            # 创建一个新的DataFrame，包含caseid和按时间戳排序的事件列表
            new_df = pd.concat([new_df, pd.DataFrame({'caseid': [current_case], 'list_of_events_ordered_by_timestamp': [events]})], ignore_index=True)
        current_case = case_id
        events = []
        # 将活动添加到案例的事件列表中
        #events.append(activity)
        # 将活动表示为一个字典并添加到案例的事件列表中
        event_dict = {'activity': activity, 'ts': timestamp}
        events.append(event_dict)

# 添加最后一个案例
new_df = pd.concat([new_df, pd.DataFrame({'caseid': [current_case], 'list_of_events_ordered_by_timestamp': [events]})], ignore_index=True)
print(new_df)
# 5. 增加持续时间和事件数量属性
new_df['duration'] = new_df['list_of_events_ordered_by_timestamp'].apply(lambda x: (x[-1]['ts'] - x[0]['ts']).total_seconds() if len(x) > 1 else 0)
new_df['event_number'] = new_df['list_of_events_ordered_by_timestamp'].apply(lambda x: len(x))

# 6. 找到持续时间最长和最短的案例
longest_case = new_df.loc[new_df['duration'].idxmax()]
shortest_case = new_df.loc[new_df['duration'].idxmin()]

# 7. 找到事件数量最多的案例（可能有多个）
max_event_number = new_df['event_number'].max()
longest_cases_by_events = new_df[new_df['event_number'] == max_event_number]

# 打印结果
print("Longest Case by Duration:")
print(longest_case)

print("\nShortest Case by Duration:")
print(shortest_case)

print("\nLongest Cases by Number of Events:")
print(longest_cases_by_events)

```
